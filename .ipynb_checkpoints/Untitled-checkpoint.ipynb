## LINEAR REGRESSION

### y = mx+c

#### y = best fit line; m = slope; x = size; c = intercept

##### if x=0, that is, if the size is 0, then y =c , or the best fit line is the intercept or a point on the y-axis

### COST FUNCTION = 1/2m sigma(i=1)(yhat-y)^2

# BOSTON HOUSE DATASET

```python
from sklearn.datasets import load_boston
```

```python
bos = load_boston()
```

```python
sns.get_dataset_names()
```

```python
df = pd.DataFrame(bos.data, columns=bos.feature_names)
```

```python
df.drop(['sex','smoker','day','time','size'],axis=1,inplace=True)
```

```python
df
```

```python
df['MEDV'] = bos.target
```

```python
X = df.drop('MEDV',axis=1).values
y = df['MEDV'].values
```

```python
X_rooms = X[:,5]
```

```python
y = y.reshape(-1,1)
X_rooms = X_rooms.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_rooms,y)
```

```python
prediction_space = np.linspace(min(X_rooms),max(X_rooms)).reshape(-1,1)
```

```python
plt.scatter(X_rooms,y,color='blue')
```

```python
plt.scatter(X_rooms,y,color='blue')
plt.plot(prediction_space, reg.predict(prediction_space), color='r', linewidth=3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
reg.fit(X_train,y_train)
```
